```python
import pandas as pd
import numpy as np
import datetime
import yfinance as yf
#import pandas.tseries.offsets.BusinessDay
from pandas.tseries.offsets import BDay
from platform import python_version
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
data=pd.read_csv("indiabulls.csv")
```

```python
data["sma50"]=data["Adj Close"].rolling(window=50).mean()
```

```python
data=data[49:].reset_index(drop=True)
```

```python
data["Date"]=data['Date'].apply(lambda x :datetime.datetime.strptime(x,"%Y-%m-%d"))
```

```python
long_position=0
sell_position=0
trade_log=pd.DataFrame(columns=["Trade_type","long_entry_time","long_exit_time","short_entry_time","short_exit_time","short_entry_price","short_exit_price","long_entry_price","long_exit_price","pnl"])

for index,today in data.iterrows():
    previous_day=data.iloc[index-1]
    
    if (previous_day['Adj Close']<previous_day['sma50'])&(today['Adj Close']>today['sma50'])&((long_position==0)or(sell_position==1)):
            ## fresh long call##
                
        if (long_position==0) & (sell_position==0) :
            long_position=1                                         
            entry_price=today['Adj Close']
            print("loop1")
            entry_time=today["Date"]
                                  
            trade_log=trade_log.append({"Trade_type":"BUY","long_entry_time":entry_time,"long_entry_price":entry_price},ignore_index=True)
         ## sell exit ###################################################  
        
        if  (sell_position==1) & (long_position==0):  
                      
             exit_price=today['Adj Close']
             print("loop2")
             exit_time=today["Date"]                      
             sell_position=0
             long_position=1
             trade_log=trade_log.append({"Trade_type":"BUY","short_exit_time":exit_time,"short_exit_price":exit_price},ignore_index=True)
             #trade_log=trade_log.append({"Trade_type":"BUY","exit_time":exit_time,"exit_price":exit_price},ignore_index=True)
                            
    if(previous_day['Adj Close']>previous_day['sma50'])&(today['Adj Close']<today['sma50'])&((long_position==1)or(sell_position==0)):
          
        ### frsh sell call ############################################################
        
        if (sell_position==0) &(long_position==0): 
              sell_position=1 
              print("loop3")
              entry_price=today["Adj Close"]
              entry_time=today["Date"]
              trade_log=trade_log.append({"Trade_type":"SELL","short_entry_time":entry_time,"short_entry_price":entry_price},ignore_index=True)
                  
        
         ## long Exit#############################################
            
        if (long_position==1)&(sell_position==0) :
             long_position=0
             sell_position=1
             print("loop4")
             exit_price=today['Adj Close']
             exit_time=today["Date"]
             trade_log=trade_log.append({"Trade_type":"SELL","long_exit_time":exit_time,"long_exit_price":exit_price},ignore_index=True)
             
          
        
```

```python
trade_log.to_csv("trade_log.csv")
```
